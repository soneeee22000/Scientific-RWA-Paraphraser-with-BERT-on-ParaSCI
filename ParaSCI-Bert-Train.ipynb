The original dataset is made up of ACl 33,981 paraphrase pairs from ACL (ParaSCI-ACL) and 316,063 pairs from arXiv (ParaSCI-arXiv) which are the gold standard paraphrase pairs meaning human annotators are typically involved. They carefully review and analyze the given text or sentence and produce alternative versions that convey the same meaning. These human-generated paraphrases are considered high-quality because they reflect the intended meaning accurately. These human annotators are the Computational Linguists who are the experts in the field of NLP and they are the ones who are responsible for creating the gold standard paraphrase pairs. 

The line count amounts to basically 338717 which is close to the combination of only the ACL and arXiv datasets -350044. 
Hence, I believe the reformatted version make use of mostly these gold standard paraphrase pairs and the rest are from S2ORC (https://github.com/allenai/s2orc) which means the dataset is already in its finest form and we can use it directly for our training/finetuning purposes. 

more about the gold standard paraphrase pairs : 

in short, gold standard paraphrases are a set of high-quality reference paraphrases that are used as a benchmark for evaluating the performance of paraphrase generation systems. They are created through a meticulous human annotation process and serve as a reliable standard for assessing the accuracy and effectiveness of automated paraphrasing algorithms.

```python
#Len represents the average number of words per sentence and Char Len represents the average number of characters per sentence.
train_df['Source_Len_Train'] = train_df['sentence1'].apply(lambda x: len(x.split()))
train_df['Target_Len_Train'] = train_df['sentence2'].apply(lambda x: len(x.split()))
train_df['Source_Char_Len_Train'] = train_df['sentence1'].apply(lambda x: len(x))
train_df['Target_Char_Len_Train'] = train_df['sentence2'].apply(lambda x: len(x))

```

```python
#let's see some examples of sentences with the highest number of words
train_df.sort_values(by=['Source_Len_Train'], ascending=False).head(10)
```

```python
#let's calculate the average number of words per sentence
train_df['Source_Len_Train'].mean()
```

```python
#how about the average number of characters per sentence
train_df['Source_Char_Len_Train'].mean()
```

```python
# to calculate the self blue score 
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.bleu_score import SmoothingFunction
smoothie = SmoothingFunction().method4

#let's calculate the self blue score
train_df['Self_Bleu_Score'] = train_df.apply(lambda x: sentence_bleu(references=[x['sentence1'].split()], hypothesis=x['sentence2'].split(), smoothing_function=smoothie), axis=1)

#let's see the first 5 examples
train_df.head()
```

```python
#the average self blue score
train_df['Self_Bleu_Score'].mean()
```

```python
#let's see the distribution of self blue score
sns.displot(train_df['Self_Bleu_Score'])
```

This distribution of the self-bleu score is good for us , actually . Since we are now sure that our sorce-target paraphrases pairs are of high quality ( i.e source and target sentences are not supposed to be the same ,but paraphrased into different words or phrase structures which means the self BLEU score is meant to be low to indicated that a sentence is written in much different way than its original input sentence ) and we can use them for our training purposes.

NOTE: Down Below as we print out some highest BLEU scores , we can see that those sentences that got 1.0 for their self BLEU score doesn't really show any changes in their target sentence at all! 

```python
#let's see some highest self blue score examples
train_df.sort_values(by=['Self_Bleu_Score'], ascending=False).head(10)
```
